```python
import os 
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
i_path = 'data/100'

data = []
for img in os.listdir(i_path):
    pic = cv2.imread(os.path.join(i_path,img))
    pic = cv2.cvtColor(pic,cv2.COLOR_BGR2RGB)
    pic = cv2.resize(pic,(80,80))
    data.append([pic])
    
np.save(os.path.join(i_path,'features'),np.array(data))
```

```python
saved = np.load(os.path.join(i_path,'features.npy'))
```
